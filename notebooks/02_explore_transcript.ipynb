# 02 — Explore Transcript

Load a transcript fixture and inspect it: view sentences with timestamps,
check stats, estimate tokens, and preview what different time windows look like.

```python
# Parameters (papermill-compatible)
book_slug = "hp-book1-ss"
chapter_index = 16  # Which chapter fixture to load
```

```python
import sys, os
# Ensure lib/ is importable regardless of CWD
_eval_root = os.path.abspath(os.path.join(os.path.dirname("__file__"), ".."))
if _eval_root not in sys.path:
    sys.path.insert(0, _eval_root)

from pathlib import Path
from lib.transcript import (
    load_fixture, load_chapters, format_transcript,
    format_transcript_display, select_window
)
from lib.token_budget import estimate_tokens
```

## Load chapter info and transcript

```python
fixture_dir = Path(f"../fixtures/{book_slug}")

chapters = load_chapters(fixture_dir / "chapters.json")
ch = chapters[chapter_index]
print(f"Chapter {ch.index}: {ch.title} ({ch.duration/60:.1f} min)\n")

sentences = load_fixture(fixture_dir / f"transcript_ch{chapter_index:02d}.json")
print(f"Loaded {len(sentences)} sentences")
print(f"Time range: {sentences[0].start_time:.1f}s — {sentences[-1].end_time:.1f}s")
```

## Full transcript with timestamps

```python
print(format_transcript_display(sentences))
```

## Stats

```python
full_text = format_transcript(sentences)
word_count = len(full_text.split())
char_count = len(full_text)
token_count = estimate_tokens(full_text)

print(f"Sentences:  {len(sentences)}")
print(f"Words:      {word_count}")
print(f"Characters: {char_count}")
print(f"Tokens (est): {token_count}")
print(f"Duration:   {sentences[-1].end_time - sentences[0].start_time:.0f}s")
print(f"")
print(f"Avg words/sentence: {word_count / len(sentences):.1f}")
print(f"Avg tokens/word:    {token_count / max(1, word_count):.2f}")
```

## Preview time windows

See what a 3-minute and 5-minute WDIM window looks like at the end of the chapter.

```python
end_pos = sentences[-1].end_time

for window_min in [3, 5, 8, 10]:
    window = select_window(sentences, position=end_pos, window_minutes=window_min)
    text = format_transcript(window)
    tokens = estimate_tokens(text)
    print(f"\n--- {window_min}-min window ({len(window)} sentences, ~{tokens} tokens) ---")
    print(text[:300])
    if len(text) > 300:
        print(f"... ({len(text) - 300} more chars)")
```
